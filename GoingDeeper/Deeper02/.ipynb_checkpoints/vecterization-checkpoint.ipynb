```python
# ML
from tensorflow.keras.datasets import reuters

# 1DCNN
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
import numpy as np
from tensorflow.keras.datasets import reuters
from tensorflow.keras.preprocessing.sequence import pad_sequences
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score
import time
```

```python
# Keras 내장 로이터 데이터 바로 로드
(x_train_raw, y_train), (x_test_raw, y_test) = reuters.load_data(
    num_words=None,    # 출현 빈도 전체 단어사용
    test_split=0.2      # 20%를 test 데이터로 분리
)

print(len(x_train), len(x_test))
```

# Q. vocab size 세가지를 어떻게 설정할 것인가?
vocab size가 너무 크면 희귀 단어까지 전부 단어로 포함시킴. 
-> embedding 크기가 커져서 느리고 과적합 위험이 올라간다.
-> 희귀 단어는 어차피 자주안나와서 모델이 잘 못 배우게 된다

vocab size가 너무 작으면
-> 많은 단어가 oov(2번 토큰)으로 통합되어서 정보가 손실되고 accuracy가 낮아진다

결론: 보통은 자주 쓰이는 단어들로 전체 토큰의 95%~99%를 커버하는 지점을 하나의 기준으로 잡는다.
      그러므로 본 실험에서는 90%(작은 vocab), 95%(중간 vocab), 99%(큰 vocab) 정도를 커버하는 vocab size를 설정하고자 


```python
from collections import Counter

# 학습셋 전체에서 토큰 빈도 세기
all_tokens = []
for seq in x_train_raw:
    all_tokens.extend(seq)

counter = Counter(all_tokens)
print("고유 토큰 수:", len(counter))
total_tokens = sum(counter.values())
print("전체 토큰 수:", total_tokens)


# coverage(커버리지) 함수 만들기 + K 탐색 함수

def find_k_for_coverage(counter, target_ratio, ignore_special=True):
    """
    target_ratio: 0.90, 0.95, 0.99 처럼 커버하고 싶은 비율
    ignore_special=True 이면 0,1,2 토큰은 제외하고 계산
    """
    # (토큰, 빈도) 리스트
    items = list(counter.items())
    
    if ignore_special:
        items = [(tok, freq) for tok, freq in items if tok >= 3]

    # 빈도 내림차순 정렬
    items.sort(key=lambda x: x[1], reverse=True)

    total = sum(freq for _, freq in items)
    cum = 0
    K = 0

    for tok, freq in items:
        cum += freq
        K += 1
        if cum / total >= target_ratio:
            break

    return K, cum / total

# 90%, 95%, 99%에 해당하는 K 찾기
targets = [0.90, 0.95, 0.99]
k_results = {}

for t in targets:
    K, cov = find_k_for_coverage(counter, t, ignore_special=True)
    k_results[t] = K
    print(f"목표 {t*100:.0f}% → K = {K}, 실제 coverage = {cov*100:.2f}%")


```

# 위의 결과에 따라 top 3000(소), 7000(중), 17000(대) 등 세개 지점으로 설정하고자 함 

```python
def coverage(counter, K):
    most_common = counter.most_common(K)
    covered = sum(freq for _, freq in most_common)
    return covered / total_tokens

for K in [3000, 7000, 17000]:
    print(f"Top {K} coverage: {coverage(counter, K)*100:.2f}%")
```

# 학습 데이터 토큰 분포
Top 3,000/7,000/17,000 단어가 각각 90%, 96%, 99%의 토큰을 커버함
작은 vocab은 정보 손실이 크지만 파라미터가 적고 빠름
큰 vocab은 정보 보존은 좋지만 희소단어도 포함시킬 가능성이 노아져 과적합 및 계산비용 증가 가능성이 있음
7000은 이 둘 사이의 절충안으로 설정

```python
# 세 가지 vocab 설정별로 데이터 만들기
vocab_settings = {
    "top_3000": 3000,   # 1️⃣ 상위 3000(소)
    "top_7000": 7000,    # 2️⃣ 상위 7000(중)
    "top_17000": 17000,  # 3️⃣ 상위 17000 (대)
}

datasets = {}

def apply_num_words(x_data, num_words):
    """
    num_words 기준으로 단어 인덱스를 잘라서
    - 인덱스 < num_words : 그대로 사용
    - 인덱스 >= num_words: OOV 토큰(2)로 치환
    """
    if num_words is None:
        return x_data  # 전체 단어 버전 그대로 쓰고 싶을 때

    new_x = []
    for seq in x_data:
        new_seq = [w if w < num_words else 2 for w in seq]  # 2 = <OOV>
        new_x.append(new_seq)
    return new_x

for name, num_words in vocab_settings.items():
    x_tr = apply_num_words(x_train_raw, num_words)
    x_te = apply_num_words(x_test_raw, num_words)
    datasets[name] = (x_tr, x_te)
    print(f"{name} 준비 완료. num_words = {num_words}")

for name, (x_tr, x_te) in datasets.items():
    print(f"{name}: train={len(x_tr)}, test={len(x_te)}")
```

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.naive_bayes import MultinomialNB, ComplementNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt
```

```python
# 모델 만들기

# 1) 정수 시퀀스를 문자열로 바꿔주는 함수 ---------------------------------
def seqs_to_texts(seqs):
    """
    정수 시퀀스 리스트를 공백으로 연결된 문자열 리스트로 변환
    예: [1, 4, 7] -> "1 4 7"
    """
    return [" ".join(map(str, seq)) for seq in seqs]

# 2) 사용할 모델들을 만들어주는 함수 ----------------------------------------
def make_models():
    models = {
        "MultinomialNB": MultinomialNB(),
        "ComplementNB": ComplementNB(),
        "LogReg": LogisticRegression(C=10000, penalty="l2", max_iter=3000),
        "LinearSVM": LinearSVC(C=1000, penalty='l1', max_iter=3000, dual=False),  # 서포트 벡터 머신
        "DecisionTree": DecisionTreeClassifier(max_depth=10, random_state=0),
        "RandomForest": RandomForestClassifier(n_estimators=5, random_state=0),
        "GradBoost": GradientBoostingClassifier(random_state=0),
        "Voting": VotingClassifier(estimators=[
                ("LogReg", LogisticRegression(C=10000, penalty="l2", max_iter=3000)),
                ("ComplementNB", ComplementNB()),
                ("GradBoost", GradientBoostingClassifier(random_state=0)),
            ],
            voting="soft",
            n_jobs=-1,
        ),
    }
    return models
```

```python
# vocab_size별로 TF-IDF 만들고, 8개 모델 accuracy 계산

# vocab_settings / datasets는 이미 위에서 만든 그대로 사용
# vocab_settings = {"top_3000": 3000, "top_7000": 7000, "top_17000": 17000}

vocab_order = ["top_3000", "top_7000", "top_17000"]
vocab_sizes = [vocab_settings[k] for k in vocab_order]   # [3000, 7000, 17000]

# 결과 저장용: model_name -> [acc_for_3000, acc_for_7000, acc_for_17000]
accuracy_results = {}
# F1 저장용: model_name -> [f1_for_3000, f1_for_7000, f1_for_17000]
f1_results = {}

for vname in vocab_order:
    num_words = vocab_settings[vname]
    x_tr_seqs, x_te_seqs = datasets[vname]

    print(f"\n=== {vname} (num_words={num_words}) 처리 중 ===")

    # 1) 정수 시퀀스를 문자열 문장으로 변환
    X_train_text = seqs_to_texts(x_tr_seqs)
    X_test_text  = seqs_to_texts(x_te_seqs)

    # 2) DTM + TF-IDF 생성
    dtmvector = CountVectorizer()
    X_train_dtm = dtmvector.fit_transform(X_train_text)
    X_test_dtm  = dtmvector.transform(X_test_text)

    tfidf_transformer = TfidfTransformer()
    tfidfv_train = tfidf_transformer.fit_transform(X_train_dtm)
    tfidfv_test  = tfidf_transformer.transform(X_test_dtm)

    # 3) 모델 생성
    models = make_models()

    # 4) 각 모델 학습 & 평가
    for mname, model in models.items():
        model.fit(tfidfv_train, y_train)
        y_pred = model.predict(tfidfv_test)
        acc = accuracy_score(y_test, y_pred)
        f1  = f1_score(y_test, y_pred, average="macro")

        # accuracy_results에 누적
        if mname not in accuracy_results:
            accuracy_results[mname] = []
        accuracy_results[mname].append(acc)

        # f1_results에 누적
        if mname not in f1_results:
            f1_results[mname] = []
        f1_results[mname].append(f1)

        print(f"{mname} @ {num_words}: accuracy = {acc:.4f}, F1(macro) = {f1:.4f}")
```

# 1DCNN

Kim's CNN 구조 (2014)

입력 시퀀스 [batch, 200]
    ↓
Embedding [batch, 200, 128]
    ↓
Conv1D (3개 병렬)
├─ filter_size=3, 100개
├─ filter_size=4, 100개  
└─ filter_size=5, 100개
    ↓
MaxPooling 
    ↓
Concatenate [batch, 300]
    ↓
Dropout (0.5)
    ↓
FC [batch, 46]

```python
# ============================================================================
# 하이퍼파라미터
# ============================================================================
PARAMS = {
    'num_words': 17000,
    'max_len': 200,
    'test_split': 0.2,
    'embedding_dim': 128,
    'filter_sizes': [3, 4, 5],
    'num_filters': 100,
    'dropout': 0.5,
    'batch_size': 64,
    'num_epochs': 20,
    'learning_rate': 0.001,
    'random_seed': 42
}

torch.manual_seed(PARAMS['random_seed'])
np.random.seed(PARAMS['random_seed'])
```

```python
# 평균 길이 확인
avg_len = np.mean([len(x) for x in x_train])
print(f"평균 길이: {avg_len:.0f}")
```

평균 길이가 200이므로 max len 200 적절함

```python
# ============================================================================
# 데이터 로드 및 전처리
# ============================================================================
(x_train, y_train), (x_test, y_test) = reuters.load_data(
    num_words=PARAMS['num_words'],
    test_split=PARAMS['test_split']
)

x_train = pad_sequences(x_train, maxlen=PARAMS['max_len'])
x_test = pad_sequences(x_test, maxlen=PARAMS['max_len'])

X_train = torch.LongTensor(x_train)
X_test = torch.LongTensor(x_test)
y_train = torch.LongTensor(y_train)
y_test = torch.LongTensor(y_test)

num_classes = len(np.unique(y_train))
print(f"Train: {len(X_train)}, Test: {len(X_test)}, Classes: {num_classes}")
```

```python
# ============================================================================
# Kim's CNN 모델
# ============================================================================
class KimCNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, num_classes, filter_sizes, num_filters, dropout):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=0)
        self.convs = nn.ModuleList([
            nn.Conv1d(embedding_dim, num_filters, fs) for fs in filter_sizes
        ])
        self.dropout = nn.Dropout(dropout)
        self.fc = nn.Linear(len(filter_sizes) * num_filters, num_classes)
    
    def forward(self, x):
        x = self.embedding(x).permute(0, 2, 1)  # [B, E, L]
        x = [torch.relu(conv(x)) for conv in self.convs]  # List of [B, F, L']
        x = [torch.max(c, dim=2)[0] for c in x]  # List of [B, F]
        x = torch.cat(x, dim=1)  # [B, F*len(filter_sizes)]
        x = self.dropout(x)
        return self.fc(x)

model = KimCNN(
    PARAMS['num_words'],
    PARAMS['embedding_dim'],
    num_classes,
    PARAMS['filter_sizes'],
    PARAMS['num_filters'],
    PARAMS['dropout']
)

print(f"Parameters: {sum(p.numel() for p in model.parameters()):,}")
```

```python
# ============================================================================
# 학습 설정
# ============================================================================
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)

train_loader = DataLoader(
    TensorDataset(X_train, y_train),
    batch_size=PARAMS['batch_size'],
    shuffle=True
)
test_loader = DataLoader(
    TensorDataset(X_test, y_test),
    batch_size=PARAMS['batch_size']
)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=PARAMS['learning_rate'], weight_decay=1e-4)
```

```python
# ============================================================================
# 학습 및 평가 함수
# ============================================================================
def train_epoch(model, loader):
    model.train()
    total_loss, correct, total = 0, 0, 0
    for X, y in loader:
        X, y = X.to(device), y.to(device)
        optimizer.zero_grad()
        output = model(X)
        loss = criterion(output, y)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
        correct += (output.argmax(1) == y).sum().item()
        total += y.size(0)
    return total_loss / len(loader), correct / total

def evaluate(model, loader):
    model.eval()
    total_loss, all_preds, all_labels = 0, [], []
    with torch.no_grad():
        for X, y in loader:
            X, y = X.to(device), y.to(device)
            output = model(X)
            total_loss += criterion(output, y).item()
            all_preds.extend(output.argmax(1).cpu().numpy())
            all_labels.extend(y.cpu().numpy())
    acc = np.mean(np.array(all_preds) == np.array(all_labels))
    f1 = f1_score(all_labels, all_preds, average='weighted')
    return total_loss / len(loader), acc, f1
```

```python
# ============================================================================
# 학습 실행
# ============================================================================
history = {'train_loss': [], 'train_acc': [], 'test_loss': [], 'test_acc': [], 'test_f1': [], 'epoch_time': []}
best_acc, best_f1 = 0, 0

print("\n학습 시작...")
total_start = time.time()

for epoch in range(PARAMS['num_epochs']):
    epoch_start = time.time()
    train_loss, train_acc = train_epoch(model, train_loader)
    test_loss, test_acc, test_f1 = evaluate(model, test_loader)
    epoch_time = time.time() - epoch_start
    
    history['train_loss'].append(train_loss)
    history['train_acc'].append(train_acc)
    history['test_loss'].append(test_loss)
    history['test_acc'].append(test_acc)
    history['test_f1'].append(test_f1)
    history['epoch_time'].append(epoch_time)
    
    if test_acc > best_acc:
        best_acc, best_f1 = test_acc, test_f1
        torch.save(model.state_dict(), 'best_model.pth')
    
    print(f"Epoch {epoch+1:2d} | Train: {train_loss:.3f}/{train_acc:.3f} | "
          f"Test: {test_loss:.3f}/{test_acc:.3f}/F1:{test_f1:.3f} | {epoch_time:.1f}s")

total_time = time.time() - total_start
print(f"\n{'='*70}")
print(f"최고 Acc: {best_acc:.4f} | 최고 F1: {best_f1:.4f} | "
      f"총 시간: {total_time:.0f}s | 평균: {np.mean(history['epoch_time']):.1f}s/epoch")
print(f"{'='*70}")
```

```python
# 시각화
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].plot(history['train_loss'], label='Train')
axes[0].plot(history['test_loss'], label='Test')
axes[0].set_title('Loss')
axes[0].legend()
axes[0].grid(alpha=0.3)

axes[1].plot(history['train_acc'], label='Train Acc')
axes[1].plot(history['test_acc'], label='Test Acc')
axes[1].plot(history['test_f1'], label='Test F1', linestyle='--')
axes[1].set_title('Accuracy & F1')
axes[1].legend()
axes[1].grid(alpha=0.3)

axes[2].plot(history['epoch_time'])
axes[2].set_title(f"Epoch Time (Avg: {np.mean(history['epoch_time']):.1f}s)")
axes[2].grid(alpha=0.3)

for ax in axes:
    ax.set_xlabel('Epoch')
    
plt.tight_layout()
plt.show()
```
