```python
def read_chatbot_data(file_path, max_samples=50000):
    pairs = []
    # BOM 대응을 위해 utf-8-sig 권장
    with open(file_path, 'r', encoding='utf-8-sig', errors='ignore') as f:
        reader = csv.DictReader(f)  # 헤더: Q, A, label
        if 'Q' not in reader.fieldnames or 'A' not in reader.fieldnames:
            raise ValueError(f"CSV에 Q/A 컬럼이 없습니다. 현재 컬럼: {reader.fieldnames}")

        for row in reader:
            q = (row.get('Q') or '').strip()
            a = (row.get('A') or '').strip()
            if not q or not a:
                continue

            # 전처리 함수가 있으면 적용
            if 'preprocess_sentence' in globals():
                q = preprocess_sentence(q)
                a = preprocess_sentence(a)

            pairs.append((q, a))
            if max_samples is not None and len(pairs) >= max_samples:
                break
    return pairs
```

```python
sp = spm.SentencePieceProcessor()
sp.Load("spm_ko_uni32k.model")
```

```python
# 전처리
import re, unicodedata

STOPWORDS = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']
_re_keep_ko_digit = re.compile(r'[^가-힣0-9\s]+')
_re_multi_space   = re.compile(r'\s+')

def preprocess_sentence(s: str) -> str:
    s = unicodedata.normalize('NFKC', '' if s is None else str(s))
    s = _re_keep_ko_digit.sub(' ', s)
    s = _re_multi_space.sub(' ', s).strip()
    return s
```

```python
# 예제 문장
sentence = "저는 지금 파이토치를 배우고 있습니다."

sentence = preprocess_sentence(sentence)
print("전처리 후의 문장:", sentence)

# 1. 토크나이징 (subword 단위로 분할)
tokens = sp.encode(sentence, out_type=str)
print("Tokenized:", tokens)

# 2. 인코딩 (서브워드를 정수 ID로 변환)
encoded = sp.encode(sentence, out_type=int)
print("Encoded:", encoded)

# 3. 디코딩 (정수 ID → 원본 문장 복원)
decoded = sp.decode(encoded)
print("Decoded:", decoded)
```

```python
# Dataset구현하기

# GPT(디코더 온리)는 단일 시퀀스만 받으므로 하나의 시퀀스로 만들고 답변 구간만 예측하는 dataset 구조로 만든다.

# 반면, transformer는 인코더-디코더(Seq2Seq)스타일이라 입력을 enc_input/dec_input/target으로 나눈다.

# GPT는 단일 입력 시퀀스만 받습니다: input_ids와 attention_mask만 주고, labels는 같은 길이로 만들어 causal shift를 모델 내부(또는 loss 계산)에서 처리합니다.
# 라벨 마스킹: 질문(Q)과 구분 토큰(<sep>), BOS는 -100으로 설정 → 손실에서 무시됨. 답변(A)와 EOS만 손실 계산.

# 패딩: PAD=0 가정(토크나이저 설정과 일치해야 함). attention_mask는 PAD=0, 나머지=1.

# <sep>: 토크나이저에 user_defined_symbols=["<sep>"]로 등록되어 있어야 합니다. 아니면 use_sep=False로 끄세요


import torch
from torch.utils.data import Dataset
# SentencePiece (sp) 객체는 외부에서 정의되어 전달된다고 가정합니다.

class ChatbotGPTDataset_MinimalChange(Dataset):
    def __init__(self, pairs, sp, max_length=40):
        super().__init__()
        self.sp = sp
        self.max_length = max_length
        self.data = []

        # GPT-style 학습을 위해 [SEP] 토큰이 필요합니다.
        # 여기서는 SentencePiece의 EOS_ID를 질문/응답 구분자로 사용하거나,
        # 별도의 토큰 ID (예: 3)를 가정해야 합니다.
        # 실제 모델 학습 시 사용한 토큰 ID를 넣어야 합니다.
        sep_id = sp.eos_id() if sp.eos_id() >= 0 else 2 # EOS를 구분자처럼 활용
        bos_id = sp.bos_id() if sp.bos_id() >= 0 else 1
        pad_id = 0 # <pad> 가정

        # 손실(Loss) 계산에서 제외할 인덱스 (Hugging Face 표준)
        IGNORE_INDEX = -100 
        
        for q_text, a_text in pairs:
            # 1) 토크나이즈
            q_ids = sp.EncodeAsIds(q_text)
            a_ids = sp.EncodeAsIds(a_text)
            
            # 2) 질문 시퀀스: [BOS] + Q + [SEP]
            # (GPT는 인코더/디코더 분리 없이 단일 시퀀스로 처리)
            q_tokens = [bos_id] + q_ids + [sep_id]
            # 응답 시퀀스: A + [EOS]
            a_tokens = a_ids + [sep_id] # 기존 코드의 [bos] + a + [eos]와 다름.

            # 3) 단일 시퀀스로 합치기 (GPT의 Input ID)
            # input_ids는 [BOS] + Q + [SEP] + A + [EOS] 로 구성됨
            full_input_ids = q_tokens + a_tokens

            # 4) 길이 제한
            if len(full_input_ids) > max_length:
                continue

            # 5) 타겟 (labels) 구성: A 시퀀스에 대해서만 손실 계산
            # 'target'은 input_ids와 같은 길이를 가지며, 
            # 질문 부분은 IGNORE_INDEX로 채워집니다.
            
            # [BOS] Q [SEP] 부분의 길이 (마스킹할 길이)
            mask_len = len(q_tokens)
            
            # 타겟 레이블은 [IGNORE_INDEX, ...] * mask_len + A_tokens 
            target_ids = [IGNORE_INDEX] * mask_len + a_tokens
            
            # 6) 고정 길이 패딩
            padding_len = max_length - len(full_input_ids)
            
            # enc_input (실제 GPT의 input_ids 역할)은 pad_id(0)로 패딩
            enc_input_padded = full_input_ids + [pad_id] * padding_len
            
            # target (실제 GPT의 labels 역할)은 IGNORE_INDEX(-100)로 패딩
            target_padded = target_ids + [IGNORE_INDEX] * padding_len
            
            # GPT 모델은 별도의 디코더 입력(dec_input)이 필요 없습니다.
            # 하지만 기존 코드 구조를 유지하기 위해, dec_input은 enc_input과 동일하게 채웁니다.
            # (이 값은 모델 학습 시 사용하지 않고 무시됩니다.)
            dec_input_padded = enc_input_padded.copy() 

            self.data.append({
                "enc_input": enc_input_padded, # GPT input_ids
                "dec_input": dec_input_padded, # (사용하지 않음)
                "target": target_padded        # GPT labels
            })

    def __len__(self):
        return len(self.data)
        
    def __getitem__(self, idx):
        sample = self.data[idx]
        
        # 기존 코드의 변수명을 그대로 사용:
        enc_input = torch.tensor(sample["enc_input"], dtype=torch.long) # Input IDs
        # dec_input은 무시되지만, 구조 유지를 위해 반환
        dec_input = torch.tensor(sample["dec_input"], dtype=torch.long) 
        target = torch.tensor(sample["target"], dtype=torch.long)       # Labels

        # GPT 모델에 맞게 변환된 텐서를 반환합니다.
        return enc_input, dec_input, target
        
```

```python
# 포지셔널 행렬
class PositionalEncoding(nn.Module):
    def __init__(self, position, d_model):
        super(PositionalEncoding, self).__init__()
        self.d_model = d_model
        self.position = position

        self.pos_encoding = self._build_pos_encoding(position, d_model)

    def _get_angles(self, position, i, d_model):
        return 1.0 / (10000.0 ** ((2.0 * (i // 2)) / d_model)) * position

    def _build_pos_encoding(self, position, d_model):
        pos = torch.arange(position, dtype=torch.float32).unsqueeze(1)
        i = torch.arange(d_model, dtype=torch.float32).unsqueeze(0)

        angle_rads = self._get_angles(pos, i, d_model)
        sines = torch.sin(angle_rads[:, 0::2])
        cosines = torch.cos(angle_rads[:, 1::2])

        pos_encoding = torch.zeros(position, d_model)
        pos_encoding[:, 0::2] = sines
        pos_encoding[:, 1::2] = cosines

        pos_encoding = pos_encoding.unsqueeze(0)  # shape: [1, position, d_model]
        return pos_encoding

    def forward(self, x):
        return x + self.pos_encoding[:, :x.size(1), :].to(x.device)
```

```python
# 스케일드 닷 프로덕트 어텐션
def scaled_dot_product_attention(query, key, value, mask=None):

    # 1) Q와 K의 내적을 통해 score(유사도) 계산
    # key.transpose(-1, -2): (batch_size, heads, depth, seq_len)
    # matmul 결과 shape: (batch_size, heads, seq_len, seq_len)
    matmul_qk = torch.matmul(query, key.transpose(-1, -2))

    # 2) depth에 따라 정규화
    depth = key.size(-1)  # depth = d_model / heads
    logits = matmul_qk / math.sqrt(depth)

    # 3) 마스크가 주어졌다면 -1e9(아주 작은 값)를 더해 소프트맥스에서 제외시키도록 함
    if mask is not None:
        # 텐서플로우: logits += (mask * -1e9)
        # 파이토치 동일 적용
        logits = logits + (mask * -1e9)

    # 4) 소프트맥스 계산해 attention weights 생성
    attention_weights = F.softmax(logits, dim=-1)

    # 5) attention weights와 value의 내적
    output = torch.matmul(attention_weights, value)

    return output, attention_weights
```

```python
# 멀티 헤드 어텐션
class MultiHeadAttention(nn.Module):
    def __init__(self, d_model, num_heads, name="multi_head_attention"):
        super(MultiHeadAttention, self).__init__()
        self.num_heads = num_heads
        self.d_model = d_model

        # d_model은 num_heads로 나누어떨어져야 함
        assert d_model % num_heads == 0

        self.depth = d_model // num_heads

        # 파이토치에서 Dense는 nn.Linear로 대응
        self.query_dense = nn.Linear(d_model, d_model)
        self.key_dense = nn.Linear(d_model, d_model)
        self.value_dense = nn.Linear(d_model, d_model)

        self.out_dense = nn.Linear(d_model, d_model)

    def split_heads(self, x, batch_size):
        """
        x: (batch_size, seq_len, d_model)
        => (batch_size, num_heads, seq_len, depth) 형태로 변환
        """
        x = x.view(batch_size, -1, self.num_heads, self.depth)
        x = x.permute(0, 2, 1, 3)  # (batch_size, num_heads, seq_len, depth)
        return x

    def forward(self, query, key, value, mask=None):
        """
        query, key, value: (batch_size, seq_len, d_model)
        mask: (batch_size, 1, seq_len, seq_len) 등으로 broadcast 가능하도록 구성
        """
        batch_size = query.size(0)

        # Q, K, V에 각각 Linear 적용
        query = self.query_dense(query)
        key = self.key_dense(key)
        value = self.value_dense(value)

        # Head 분할
        query = self.split_heads(query, batch_size)
        key = self.split_heads(key, batch_size)
        value = self.split_heads(value, batch_size)

        # 스케일드 닷 프로덕트 어텐션
        scaled_attention, _ = scaled_dot_product_attention(query, key, value, mask)

        # (batch_size, num_heads, seq_len, depth) -> (batch_size, seq_len, num_heads, depth)
        scaled_attention = scaled_attention.permute(0, 2, 1, 3).contiguous()

        # 다시 (batch_size, seq_len, d_model)로 합치기
        concat_attention = scaled_attention.view(batch_size, -1, self.d_model)

        # 최종 Dense
        output = self.out_dense(concat_attention)
        return output
```

```python
# 패딩 마스킹
def create_padding_mask(x):
    # x == 0 위치를 찾아 float형 1로 변환
    mask = (x == 0).float()
    # (batch_size, seq_len) -> (batch_size, 1, 1, seq_len)
    mask = mask.unsqueeze(1).unsqueeze(2)
    return mask
```

```python
# 룩 어헤드 마스킹
def create_look_ahead_mask(x):
    seq_len = x.size(1)

    # (seq_len, seq_len) 크기의 하삼각 행렬(tril) 생성 후 1에서 빼서
    # 상삼각이 1, 하삼각(자기 자신 포함)이 0이 되도록 설정
    # => 미래 토큰(자신 인덱스보다 큰 위치) 마스킹
    look_ahead_mask = 1 - torch.tril(torch.ones((seq_len, seq_len)))

    # 패딩 마스크 생성 (shape: (batch_size, 1, 1, seq_len))
    padding_mask = create_padding_mask(x)

    # look_ahead_mask: (seq_len, seq_len) -> (1, seq_len, seq_len)
    look_ahead_mask = look_ahead_mask.unsqueeze(0)
    # -> (1, seq_len, seq_len) -> (1, 1, seq_len, seq_len)
    look_ahead_mask = look_ahead_mask.unsqueeze(1)
    look_ahead_mask = look_ahead_mask.to(x.device)

    # look-ahead 마스크와 패딩 마스크를 합성 (둘 중 하나라도 1이면 마스킹)
    # 최종 shape은 브로드캐스팅으로 (batch_size, 1, seq_len, seq_len)
    combined_mask = torch.max(look_ahead_mask, padding_mask)
    return combined_mask
```

```python
# 디코더층(두개의 서브층=masked multihead self attention + ffnn)

# 디코더 층에서 transformer와 GPT의 차이점은 gpt는 인코더-디코더 어텐션이 없는 2개의 서브층으로 구성됨
# 활성화 함수는 GELU 사용

import torch
import torch.nn as nn
import torch.nn.functional as F

# 💡 [필수 가정] MultiHeadAttention 클래스는 외부에 정의되어 있다고 가정합니다.

class DecoderLayer(nn.Module):
    def __init__(self, d_model, num_heads, ff_dim, dropout=0.1):
        super(DecoderLayer, self).__init__()

        # 첫 번째 서브 레이어 (Masked Multi-Head Self Attention)
        self.self_mha = MultiHeadAttention(d_model, num_heads)
        self.norm1 = nn.LayerNorm(d_model, eps=1e-6)

        # 인코더-디코더 어텐션 관련 모듈은 GPT에서 제거됨

        # 피드포워드 네트워크 (FFNN)
        self.ffn = nn.Sequential(
            nn.Linear(d_model, ff_dim),
            nn.GELU(),                   # ReLU 대신 GPT 표준인 GELU 사용
            nn.Linear(ff_dim, d_model)
        )
        self.norm2 = nn.LayerNorm(d_model, eps=1e-6) # LayerNorm 재정의

        self.dropout1 = nn.Dropout(dropout)
        self.dropout2 = nn.Dropout(dropout)

    # 💡 forward 시그니처에서 enc_outputs와 padding_mask 제거
    def forward(self, x, look_ahead_mask=None): 
        
        # 1) Masked Self-Attention
        self_attn_out = self.self_mha(x, x, x, mask=look_ahead_mask)
        self_attn_out = self.dropout1(self_attn_out)
        out1 = self.norm1(x + self_attn_out)

        # 2) Cross-Attention 단계는 GPT에서 제거되므로 out2는 out1과 같습니다.
        out2 = out1

        # 3) Feed-Forward Network
        ffn_out = self.ffn(out2)
        ffn_out = self.dropout2(ffn_out)
        out3 = self.norm2(out2 + ffn_out)

        return out3
```

```python
# 디코더층 쌓기

import math

class Decoder(nn.Module):
    def __init__(self,
                 vocab_size,
                 num_layers,
                 ff_dim,
                 d_model,
                 num_heads,
                 dropout=0.1):
        super(Decoder, self).__init__()
        self.d_model = d_model

        self.embedding = nn.Embedding(vocab_size, d_model)
        # 💡 [필수 가정] PositionalEncoding 클래스는 외부에 정의되어 있습니다.
        self.pos_encoding = PositionalEncoding(position=vocab_size, d_model=d_model) 
        self.dropout = nn.Dropout(dropout)

        # DecoderLayer 쌓기
        self.dec_layers = nn.ModuleList([
            DecoderLayer(d_model, num_heads, ff_dim, dropout)
            for _ in range(num_layers)
        ])

    # 💡 [핵심 수정] enc_outputs와 padding_mask 인수를 제거
    def forward(self, x, look_ahead_mask=None): 
        
        x = self.embedding(x) * math.sqrt(self.d_model)
        x = self.pos_encoding(x)
        x = self.dropout(x)

        # num_layers만큼 쌓인 DecoderLayer 통과
        for layer in self.dec_layers:
            # 💡 [핵심 수정] 수정된 DecoderLayer 호출에 맞게 look_ahead_mask만 전달
            x = layer(x, look_ahead_mask) 

        return x
```

```python
# GPT1 함수 정의

class GPT1(nn.Module):
    def __init__(self, vocab_size, num_layers, units, d_model, num_heads, dropout=0.1):
        super(GPT1, self).__init__()

        # 인코더는 제거됨 (주석 처리 상태 유지)

        # 디코더 (GPT의 핵심 블록)
        self.decoder = Decoder(
            vocab_size=vocab_size, num_layers=num_layers, ff_dim=units, 
            d_model=d_model, num_heads=num_heads, dropout=dropout
        )

        # 최종 출력층
        self.final_linear = nn.Linear(d_model, vocab_size)

    # inputs = enc_input 역할 (무시됨), dec_inputs = 실제 GPT 입력 시퀀스
    def forward(self, inputs, dec_inputs): 
        # 1. inputs (enc_input 역할)은 무시됩니다.
        
        # 2. Causal Attention 마스크 생성
        look_ahead_mask = create_look_ahead_mask(dec_inputs) 

        # 3. dec_padding_mask 생성 로직 제거
        # dec_padding_mask = create_padding_mask(inputs) -> 제거됨
        dec_padding_mask = None # Decoder가 받지 않으므로 사용 안 함

        # 4. 디코더 수행
        dec_outputs = self.decoder(
            x=dec_inputs,           # 실제 입력 시퀀스
            look_ahead_mask=look_ahead_mask,
            # padding_mask=dec_padding_mask <-- Decoder가 받지 않으므로 전달 안 함
        )

        # 5. 최종 Dense
        logits = self.final_linear(dec_outputs)
        
        return logits
```

```python
# 하이퍼파라미터 설정
NUM_LAYERS = 2     # GPT의 디코더 층 수 (기존 인코더/디코더 층 수 역할)
D_MODEL = 256      # 임베딩 및 내부 표현 차원
NUM_HEADS = 8      # 멀티헤드 어텐션에서의 헤드 수
UNITS = 512        # 피드포워드 신경망의 은닉 차원 (GPT의 feedforward dimension 역할)
DROPOUT = 0.1      # 드롭아웃 비율
VOCAB_SIZE = 8000  # 단어 집합 크기(예시)

#---------------------------------------
# 모델 생성 (수정된 부분)
# ----------------------------------------------------
# 'Transformer' 대신 GPT에 적합한 클래스 이름으로 대체
model = GPT1( 
    vocab_size=VOCAB_SIZE,
    num_layers=NUM_LAYERS, # 이 값이 디코더 층 수가 됩니다.
    units=UNITS,
    d_model=D_MODEL,
    num_heads=NUM_HEADS,
    dropout=DROPOUT
)

print("-" * 30)
print("생성된 모델 구조:")
print(model)
```

```python
# 커스텀된 학습률
def get_lr_lambda(d_model, warmup_steps=4000):
    d_model = float(d_model)
    def lr_lambda(step):
        # step은 0부터 시작하므로 +1로 보정
        step = step + 1
        return (d_model ** -0.5) * min(step ** -0.5, step * (warmup_steps ** -1.5))
    return lr_lambda
```

```python
# 훈련하기

import torch

def train_step(model, batch, optimizer, loss_function, device):
    model.train()
    
    # 1. 텐서 할당 및 GPU 이동 (타입 명시 추가)
    # [최소 수정] 리스트 컴프리헨션 대신 분리하여, -100이 포함된 텐서도 안전하게 long 타입으로 이동
    enc_input, dec_input, target = batch
    
    # enc_input (input_ids)과 dec_input (dummy)
    enc_input = enc_input.to(device, dtype=torch.long)
    dec_input = dec_input.to(device, dtype=torch.long) 
    
    # target (labels)은 -100이 포함되어 있으므로 long 타입 강제
    labels = target.to(device, dtype=torch.long) # GPT의 실제 타겟은 'labels' 역할을 함

    optimizer.zero_grad()

    # 모델 포워드 패스 (모델의 인자 구조를 유지)
    logits = model(enc_input, dec_input)  # (batch_size, seq_len, vocab_size)

    # 2. Loss 계산 (타겟 변수명 수정)
    # 💡 [필수 수정] 타겟 변수를 'target'이 아닌 'labels'로 변경
    loss = loss_function(logits.permute(0, 2, 1), labels)  # (batch_size, vocab_size, seq_len) 필요

    # Backpropagation
    loss.backward()
    optimizer.step()

    # 3. 정확도 계산 (마스킹 ID 수정)
    # 💡 [필수 수정] pad_id를 GPT 마스킹 값인 -100으로 변경
    # (sp.pad_id()는 보통 0이며, -100 마스킹을 무시하지 못함)
    return loss.item(), accuracy_function(logits, labels, pad_id=-100)
```

```python
def train(model, dataloader, optimizer, loss_function, scheduler, num_epochs, device):
    model.to(device)

    for epoch in range(num_epochs):
        total_loss, total_acc = 0, 0
        for step, batch in enumerate(dataloader):
            loss, acc = train_step(model, batch, optimizer, loss_function, device)
            total_loss += loss
            total_acc += acc

            # 일정 스텝마다 로그 출력
            if step % 100 == 0:
                print(f"[Epoch {epoch+1}, Step {step}] Loss: {loss:.4f}, Acc: {acc:.4f}")

            # 학습률 스케줄러 업데이트
            scheduler.step()

        avg_loss = total_loss / len(dataloader)
        avg_acc = total_acc / len(dataloader)
        print(f"Epoch {epoch+1} Completed - Avg Loss: {avg_loss:.4f}, Avg Acc: {avg_acc:.4f}")
```

```python
%%time

train(
    model=model,
    dataloader=dataloader,
    optimizer=optimizer,
    loss_function=loss_function,
    scheduler=scheduler,
    num_epochs=50,  # 원하는 에폭 수
    device=device
)
```

```python
import torch

def decoder_inference(model, sentence, tokenizer, device='cpu'):
    START_TOKEN = tokenizer.bos_id()
    END_TOKEN = tokenizer.eos_id()
    MAX_LENGTH = 40
    # 💡 [필수 가정] 데이터셋에서 사용한 [SEP] ID를 정의합니다. (일반적으로 3 또는 토크나이저에서 가져옴)
    SEP_TOKEN = 3 

    # 전처리 (preprocess_sentence 함수는 정의되어 있다고 가정)
    # sentence = preprocess_sentence(sentence) 

    # 1. GPT 입력 시퀀스 구성: [BOS] + Q + [SEP]
    q_ids = tokenizer.encode(sentence)
    # 💡 [핵심 수정 1] enc_input_ids 대신 Q + SEP로 통합
    input_tokens = [START_TOKEN] + q_ids + [SEP_TOKEN]
    
    # 💡 [핵심 수정 2] enc_input 대신 input_ids를 통합 시퀀스로 사용
    input_ids = torch.tensor([input_tokens], dtype=torch.long, device=device)

    # 💡 [수정] dec_input은 더 이상 필요하지 않습니다.
    # dec_input = torch.tensor([[START_TOKEN]], dtype=torch.long, device=device) # 주석 처리

    model.eval()  # 모델 평가 모드
    with torch.no_grad():
        for i in range(MAX_LENGTH):
            # 💡 [핵심 수정 3] 모델 forward: input_ids를 두 번 전달 (모델 시그니처 유지)
            # model(enc_input, dec_input) -> model(input_ids, input_ids)
            # GPT1 모델은 두 번째 인자를 무시해야 함
            logits = model(input_ids, input_ids) 

            # 마지막 타임스텝의 예측만 추출: shape (1, vocab_size)
            last_step_logits = logits[:, -1, :]

            # argmax로 가장 높은 확률의 토큰 선택
            predicted_id = torch.argmax(last_step_logits, dim=-1)  # shape: (1,)

            # 종료 토큰이면 중단
            if predicted_id.item() == END_TOKEN:
                break

            # 💡 [핵심 수정 4] dec_input 대신 input_ids에 예측 토큰을 이어붙임
            predicted_id_tensor = predicted_id.unsqueeze(0)  # shape (1,1)
            input_ids = torch.cat([input_ids, predicted_id_tensor], dim=1) # input_ids 확장

    # 최종 시퀀스: input_ids (1, seq_len)에서 (seq_len,)로
    # Q와 A가 모두 포함된 전체 시퀀스를 반환
    output_sequence = input_ids.squeeze(0).tolist() 

    return output_sequence
```

```python
def sentence_generation(model, sentence, tokenizer, device='cpu'):
    # 디코더 인퍼런스 -> 예측된 토큰 시퀀스
    output_seq = decoder_inference(model, sentence, tokenizer, device=device)

    # 토크나이저로 디코딩 (패딩, START/END 토큰 등은 제외하거나 처리)
    # 여기서는 단순히 tokenizer.decode() 직접 호출
    predicted_sentence = tokenizer.decode(
        [token for token in output_seq if token < tokenizer.GetPieceSize()]
    )

    print("입력 :", sentence)
    print("출력 :", predicted_sentence)
    return predicted_sentence
```
